## Set Font Properties for Title and Labels
You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

```python
xpt=np.array([23,45,23,56])
ypt=np.array([13,45,67,89])
plt.plot(xpt,ypt,color='red')
font1 = {'family':'serif','color':'blue','size':15}
font2 = {'family':'serif','color':'darkred','size':15}
font3 = {'family':'serif','color':'red','size':24}
plt.xlabel("x axis data",fontdict=font1)
plt.ylabel("y axis data",fontdict=font2)
plt.title("FONT SIZE",fontdict=font3)
plt.show()
```

```python
x=np.array([34,56,12,36,37,89])
plt.pie(x)
plt.show()
```

By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:

### Note:

 The size of each wedge is determined by comparing the value with all the other values, by using this formula:

The value divided by the sum of all values: x/sum(x)


```python
x=np.array([34,56,12,36,37,89])
mylabels=('Telagana','Tamil Nadu','Sikkim','Karnataka','Madhya Pradesh','Rajasthan')
plt.pie(x,labels=mylabels)
plt.show()
```

### matplotlib.pyplot.figure() Function:
The figure() function in pyplot module of matplotlib library is used to create a new figure.


Syntax: matplotlib.pyplot.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True, FigureClass=, clear=False, 

**kwargs)
Parameters: This method accept the following parameters that are described below:

1) num : This parameter is provided, and a figure with this id already exists.

2) figsize(float, float): These parameter are the width, height in inches.

4) facecolor : This parameter is the background color.

5) edgecolor : This parameter is the border color.

6) frameon : This parameter suppress drawing the figure frame.

7) FigureClass : This parameter use a custom Figure instance.


8) Clear : This parameter if True and the figure already exists, then it is cleared.

9) Return value: This method returns the following values:

figure : This returns the Figure instance returned will also be passed to new_figure_manager in the backends.

#### The axisartist toolkit
The axisartist contains a custom Axes class that is meant to support curvilinear grids (e.g., the world coordinate system in astronomy). Unlike Matplotlib's original Axes class which uses Axes.xaxis and Axes.yaxis to draw ticks, ticklines, etc., axisartist uses a special artist (AxisArtist) that can handle ticks, ticklines, etc. for curved coordinate systems.

axisartist
The axisartist module provides a custom (and very experimental) Axes class, where each axis (left, right, top, and bottom) have a separate associated artist which is responsible for drawing the axis-line, ticks, ticklabels, and labels. You can also create your own axis, which can pass through a fixed position in the axes coordinate, or a fixed position in the data coordinate (i.e., the axis floats around when viewlimit changes).

The axes class, by default, has its xaxis and yaxis invisible, and has 4 additional artists which are responsible for drawing the 4 axis spines in "left", "right", "bottom", and "top". They are accessed as ax.axis["left"], ax.axis["right"], and so on, i.e., ax.axis is a dictionary that contains artists (note that ax.axis is still a callable method and it behaves as an original Axes.axis method in Matplotlib).

```python
# To create an Axes,
import mpl_toolkits.axisartist as AA
from mpl_toolkits import mplot3d
fig = plt.figure()
fig.add_axes([0.1, 0.1, 0.8, 0.8], axes_class=AA.Axes)
```

# Patches:- 
               refer to graphical objects that can be added to a plot to represent various shapes such as rectangles,             circles, polygons, etc. They are part of the matplotlib.patches module.

Patches are useful for visually highlighting or annotating specific regions or elements on a plot. They can be customized with different colors, transparency, linewidths, and other visual properties to suit specific needs.

Some commonly used patch classes in Matplotlib include:

**Rectangle:** Represents a rectangle with given width, height, and coordinates.

**Circle:** Represents a circle with a given radius and center coordinates.

**Polygon:** Represents a polygon with a given list of vertices.

**Ellipse:** Represents an ellipse with given width, height, and center coordinates.

**Arrow:** Represents an arrow with a given start point, end point, and arrowhead properties.

To use patches in Matplotlib, you need to import the patches module from matplotlib and create instances of the desired patch classes. These patch objects can then be added to a plot using the add_patch method of an Axes object.

Here's an example of using patches to draw a rectangle and a circle on a plot:

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots()

# Create a rectangle patch
rect = patches.Rectangle((0.3, 0.5), 0.5, 0.7, edgecolor='red', facecolor='blue', alpha=0.5)

# Create a circle patch
circle = patches.Circle((0.8, 0.5), 0.2, edgecolor='green', facecolor='none', linestyle='dashed')

# Add the patches to the plot
ax.add_patch(rect)
ax.add_patch(circle)

# Set plot limits and labels
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Display the plot
plt.show()
```
